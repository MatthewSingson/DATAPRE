## Scraping p.97

```python
import requests
import pandas as pd
URL="https://www.manilatimes.net/news/page/97"
```

```python
page=requests.get(URL)
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(page.content, 'html.parser')
```

```python
#id = news_main is where the articles are stored
news_content=soup.find("div", {"class": "col-md-9 col-xs-12"})

news_item=news_content.find_all("div", {"class": "col-content-1"})
author = news_content.find_all('a')
title = news_content.find_all('a')
article = news_content.find_all('a')
news_article = [tag['href'] for tag in article if 'href' in tag.attrs]
news_author = [tag['title'] for tag in author if 'title' in tag.attrs]
news_title = [tag['title'] for tag in title if 'title' in tag.attrs]
date = news_content.find_all("span", {"class": "roboto-a"})
```

```python
extracted_news_info = []
news_title_start = 16
news_author_start = 17
news_date_start = 1
news_article_start = 16

info_row=news_content.content

for count in news_item:
    testdte = date[news_date_start].text.replace(" ", "").split('\n')
    
    if testdte[2] == "March11,2022":
        dte = testdte[2]
        ttle = news_title[news_title_start]
        fullart = news_article[news_article_start]
        athr = news_author[news_author_start]
        news_date_start += 1
        news_title_start += 3
        news_author_start += 3
        news_article_start += 3
        extracted_news_info.append((dte, ttle, fullart, athr))
    
    elif testdte[2] == "March12,2022":
        dte = testdte[2]
        ttle = news_title[news_title_start]
        fullart = news_article[news_article_start]
        athr = news_author[news_author_start]
        news_date_start += 1
        news_title_start += 3
        news_author_start += 3
        news_article_start += 3
        extracted_news_info.append((dte, ttle, fullart, athr))
    else:
        news_date_start += 1
        news_title_start += 3
        news_author_start += 3
        news_article_start += 3
```

```python
df_news_list_1 = pd.DataFrame(extracted_news_info)
df_news_list_1.columns = ['Date', 'Title', 'Full Article', 'Author']
df_news_list_1
```
